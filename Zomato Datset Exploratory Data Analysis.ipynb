```python
df_country = pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#Merging two dataset based on the commonality of 'Country Code' column
final_df= pd.merge(df,df_country,on = 'Country Code',how = 'left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
#pie chart-- Top 3 countries that use zomato
plt.pie(country_val[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

Observation: Zomatos's maximum records or transactions are India after that United States then United Kingdom

```python
final_df.columns
```

```python
#It will create a dataframe by groupby object
final_df.groupby(['Aggregate rating','Rating color', 'Rating text'])
```

```python
# It will show the size of the dataframe created by the groupby object
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
# This will show the dataframe as index
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
```

```python
# The rename function will create 'Rating Count' column where the coulmn is 0 
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations
1. When Ratings are Between 4.5 to 4.9 --> Excelent
2. When Ratings are between 4.0 to 4.4 --> Very Good
3. When Ratings are  between 3.5 to 3.9 --> Good
4. When Ratings are between 2.5 to 3.4 --> Average
5. When Ratings are between 1.8 to2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6) #It will increase the size of the barchart
sns.barplot(x='Aggregate rating', y ='Rating Count', data = ratings) #It will plot the bar chart with the help of seaborn
```

```python

sns.barplot(x='Aggregate rating', y ='Rating Count', hue = 'Rating color',data = ratings,palette = ['blue','red','orange','yellow','green','green']) #It will plot the bar chart with the help of seaborn
```

## Observations
1. Nat Rated Count is very high
2. Maximum number of ratings are from 2.5 to 3.9

```python
# Count plot
sns.countplot(x='Rating color',data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```
